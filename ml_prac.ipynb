## ***Supervised Learning and Regression***

Supervised Learning is a type of machine learning used to train models from labelled training data. It allows you to predict the output for future or unseen data.

Labelled data is a group of samples or features that are tagged with one or more labels.

Training data need to be fed to a machine learning algorithm in order for it to generate a machine learning model that learns from your algorithms. The learning algorithm discovers patterns in the training data such that the input parameters correspond to the target. The eoutput of the process is a  machine learning model that you can then use to make predictions. This process is also called **Learning**.

The training data consists of inputs also knownm as features and output also known as labels. Eg: Color, Size and Weight are features of fruit and its name will be its label.


There are two steps involved in the building of a supervised learning algorithm:
